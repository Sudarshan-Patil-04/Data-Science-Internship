**Dataset Overview**

```python
import pandas as pd

df = pd.read_csv("cleaned_employees.csv")

df.head()
df.shape
df.info()
```

```python
df.describe()
```

**Data Visualization**

Histogram

```python
import matplotlib.pyplot as plt

df.hist()
plt.show()
```

Bar Chart

```python
df['First Name'].value_counts().plot(kind='bar')
plt.title("Employee Count by First Name")
plt.xlabel("First Name")
plt.ylabel("Count")
plt.show()
```

Correlation Matrix

```python
corr = df.corr(numeric_only=True)

plt.imshow(corr)
plt.colorbar()
plt.xticks(range(len(corr.columns)), corr.columns, rotation=90)
plt.yticks(range(len(corr.columns)), corr.columns)
plt.title("Correlation Matrix")
plt.show()
```

Scatter Plot

```python
plt.scatter(df['Bonus %'], df['Salary'])
plt.xlabel("Bonus %")
plt.ylabel("Salary")
plt.title("Bonus % vs Salary")
plt.show()
```
