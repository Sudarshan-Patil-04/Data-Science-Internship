**1. Import Libraries**

```python
import pandas as pd
import numpy as np
```

**2. Load Dataset**

```python
df = pd.read_csv('/content/employees.csv')

print("Original Dataset Shape:", df.shape)
df.head()
```

**3. Check Data Info**

```python
df.info()

print("\nMissing Values:\n", df.isnull().sum())

print("\nDuplicate Rows:", df.duplicated().sum())
```

4. Handle Missing Values

```python
df['Team'].fillna('Unknown', inplace=True)
df['Gender'].fillna(df['Gender'].mode()[0], inplace=True)


df['Senior Management'].fillna(False, inplace=True)


df['Salary'].fillna(df['Salary'].median(), inplace=True)
df['Bonus %'].fillna(df['Bonus %'].mean(), inplace=True)

print("Missing values handled.")
```

5. Remove Duplicates

```python
before = df.shape[0]

df.drop_duplicates(inplace=True)

after = df.shape[0]

print(f"Removed {before - after} duplicate rows")
```

6. Fix Data Formats

```python
df['Start Date'] = pd.to_datetime(df['Start Date'], errors='coerce')


df['First Name'] = df['First Name'].str.strip().str.title()
df['Team'] = df['Team'].str.strip().str.title()
df['Gender'] = df['Gender'].str.capitalize()


df['Salary'] = pd.to_numeric(df['Salary'], errors='coerce')
df['Bonus %'] = pd.to_numeric(df['Bonus %'], errors='coerce')

print("Data formats corrected.")
```

```python
print("\nRemaining Missing Values:\n", df.isnull().sum())

print("\nFinal Shape:", df.shape)
```
